## Exercício 06
Elabore um programa que recebe o seu nome, endereço e hobby e mostra cada uma das informações da seguinte forma:
- Nome -> Letra maiúscula
- Endereço -> Letra minúscula
- Hobby -> Primeira letra maiúscula

**Exemplo Entrada:**
```
Nome: bruno fabri
Endereço: Rua ABC
Hobby: jogar cs
```

**Exemplo Saída:**
```
Nome: BRUNO FABRI
Endereço: rua abc
Hobby: Jogar cs
```

````python
entnome = input('Nome: ')
entadress = input('Endereço: ')
enthobby = input('Hobby: ')

nome = 'Nome:'
adress = 'Endereço:'
hobby = 'Hobby:'

print(nome, entnome.upper())
print(adress, entadress.lower())
print(hobby, enthobby.title())
````
